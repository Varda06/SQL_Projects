### Connecting to database

```python
#connect to a database so that I can use SQL
conn = sqlite3.connect(name) #if the db does not exist, this creates a Any_Database_Name.db file in the current directory
#store your table in the database:
df.to_sql("table_4", conn)
```

```python
#Top Cities by Registration
get_city_split = ''' 
Select City , Count(City) as city_split
from table_4
GROUP BY 1
ORDER BY 2 DESC
LIMIT 10
''' 
get_city_split_df = pd.read_sql(get_city_split, conn)
get_city_split_df
```

```python
#EV Adoption over years
get_ev_adoption_split = ''' 
Select model_year , Count(*) as ev_adoption
from table_4
GROUP BY 1
ORDER BY 2 DESC
''' 
get_ev_adoption_split_df = pd.read_sql(get_ev_adoption_split, conn)
get_ev_adoption_split_df
```

```python
#Ev Split type Split
get_ev_type_split = ''' 
Select Electric_Vehicle_Type , Count(Electric_Vehicle_Type) as ev_split
from table_4
GROUP BY 1
ORDER BY 2 DESC
''' 
get_ev_type_split_df = pd.read_sql(get_ev_type_split, conn)
get_ev_type_split_df
```

```python
#Top 10 makes by Vehicle type
get_leaders_type_split = ''' 
WITH CTE AS
(
Select  Make , Electric_Vehicle_Type , Count(Electric_Vehicle_Type) as ev_split,
DENSE_RANK() OVER(PARTITION BY  Electric_Vehicle_Type ORDER BY Count(Electric_Vehicle_Type) DESC ) as rnk
from table_4
GROUP BY 1 , 2
ORDER BY 3 DESC
)
select Make , Electric_Vehicle_Type , ev_split
from  cte
where rnk <= 10
order by 2
''' 
get_leaders_type_split_df = pd.read_sql(get_leaders_type_split, conn)
get_leaders_type_split_df
```

```python
#Split by make
get_make_split = ''' 
Select make , Count(make) as make_split
from table_4
GROUP BY 1
ORDER BY 2 DESC
limit 10
''' 
get_make_split_df = pd.read_sql(get_make_split, conn)
get_make_split_df
```

```python
#Top 10 EV vehicle registrations
get_model_split = ''' 
Select make,  model,  Count(model) as model_split
from table_4
GROUP BY 1 , 2
ORDER BY 3 DESC
LIMIT 10
''' 
get_model_split_df = pd.read_sql(get_model_split, conn)
get_model_split_df
```

```python
#Top Vehicle registrations for each make by make ,  model and  year
get_year_split = ''' 
WITH CTE AS (
    SELECT make, model, model_year,  COUNT(*) AS registrations,
           DENSE_RANK() OVER(PARTITION BY make ORDER BY COUNT(*) DESC) AS rnk
    FROM table_4
    GROUP BY make, model , model_year
)
SELECT make, model, model_year, registrations
FROM CTE
WHERE rnk = 1
ORDER BY registrations DESC
''' 
get_year_split_df = pd.read_sql(get_year_split, conn)
get_year_split_df
```

```python
#Split by Clean_Alternative_Fuel_Vehicle_(CAFV)_Eligibility by make and model
get_CAFV_cnt_split = ''' 
Select "Clean_Alternative_Fuel_Vehicle_(CAFV)_Eligibility", COUNT("Clean_Alternative_Fuel_Vehicle_(CAFV)_Eligibility") as CAFV_cnt_split
from table_4
GROUP BY 1
ORDER BY 2 DESC
''' 
get_CAFV_cnt_split_df = pd.read_sql(get_CAFV_cnt_split, conn)
get_CAFV_cnt_split_df
```

```python
#Split by Clean_Alternative_Fuel_Vehicle_(CAFV)_Eligibility by make and model
get_CAFV_split = ''' 
Select make , model , "Clean_Alternative_Fuel_Vehicle_(CAFV)_Eligibility", COUNT("Clean_Alternative_Fuel_Vehicle_(CAFV)_Eligibility") as CAFV_split
from table_4
GROUP BY 1
ORDER BY 4 DESC
LIMIT 20 
''' 
get_CAFV_split_df = pd.read_sql(get_CAFV_split, conn)
get_CAFV_split_df
```

```python
#Average Electric Range by make
get_avg_range_split = ''' 
Select make, ROUND(Avg(Electric_Range),2) as average_electric_Range
from table_4
GROUP BY 1 
ORDER BY 2 DESC
LIMIT 10
''' 
get_avg_range_split_df = pd.read_sql(get_avg_range_split, conn)
get_avg_range_split_df
```

```python
# TOP Electric Range by make and model

get_top_range_split = ''' 
Select make,  model, Electric_Range as top_electric_Range
from table_4
GROUP BY 1, 2
ORDER BY 3 DESC
LIMIT 10
''' 
get_top_range_split_df = pd.read_sql(get_top_range_split, conn)
get_top_range_split_df
```

```python
#Check Missing values
get_miss_df = '''
SELECT COUNT(*) AS missing_prices
FROM table_4
WHERE Base_MSRP IS NULL OR Base_MSRP = 0
'''

get_miss_split_df = pd.read_sql(get_miss_df, conn)
get_miss_split_df
```

```python
#Group by Make: Average Price & Range by Manufacturer
get_price_range_split = ''' 
SELECT Make, 
       ROUND(AVG(Base_MSRP), 2) AS avg_price, 
       ROUND(AVG(Electric_Range), 2) AS avg_range
FROM table_4
WHERE Base_MSRP > 0
GROUP BY Make
ORDER BY avg_range DESC
''' 
get_price_range_split_df = pd.read_sql(get_price_range_split, conn)
get_price_range_split_df
```

## Key Insights from the Electric Vehicle (EV) Analysis:
1) Top Cities by EV Registrations:
Seattle leads with 35,664 registrations, followed by Bellevue (10,966) and Vancouver (8,103). Urban centers and tech hubs dominate EV adoption.

2) EV Adoption Over Time:
2023 saw the highest registrations (59,060), indicating a recent surge in EV adoption. Data for 2024–2025 (future years) may reflect pre-registrations or data entry errors.

3) EV Type Split:
Battery Electric Vehicles (BEVs) dominate (177,151), while Plug-in Hybrids (PHEVs) account for only 46,844 registrations. This aligns with global trends favoring fully electric vehicles. Within BEV, Tesla is the leader and within PHEV, Toyota is leading

4) Vechicle registrations by each make:
Analyzed the model most registered for each make along with the year to see the most bought model within each make

5) Manufacturer Dominance:
Tesla leads with 96,180 registrations, followed by Chevrolet (16,405) and Nissan (15,259). Tesla’s Model Y (46,583) and Model 3 (34,462) are the most popular models.

6) CAFV Eligibility:
130,442 entries have unknown eligibility due to unresearched battery ranges. Only 71,438 are confirmed eligible, highlighting potential gaps in data quality.

7) Electric Range Analysis:
Jaguar has the highest average range (196.84 miles), while Tesla’s average (62.51 miles) seems unusually low, possibly due to older models or data inconsistencies.
The Porsche Macan has the highest single-model range (308 miles).
